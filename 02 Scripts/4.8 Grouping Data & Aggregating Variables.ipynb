# Create a flag for low spender and high spender

```python
# Create column for average spender in order to find lower and higher spending data
df_ords_prods_merge['average_spending'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check 
df_ords_prods_merge.head(15)
```

```python
# Check columns that need focus for easier comprehension
df_ords_prods_merge[['user_id', 'average_spending']].head(15)
```

```python
# Create flags for lower and higher spending
df_ords_prods_merge.loc[df_ords_prods_merge['average_spending'] >= 10, 'spending_flag'] = 'High spender'
df_ords_prods_merge.loc[df_ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Use value_counts()
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check
df_ords_prods_merge[['user_id', 'average_spending', 'spending_flag']].head(15)
```

# Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median.

```python
# Create column for order frequency flags 
df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Check
df_ords_prods_merge.head(20)
```

```python
#Check with columns we need to focus on
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'order_frequency']].head(15)
```

```python
# Create flags for order frequency 
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] <= 20) & (df_ords_prods_merge['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Use value_counts()
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'order_frequency', 'order_frequency_flag']].head(15)
```
